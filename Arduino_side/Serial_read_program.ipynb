```python
from itertools import count
import serial
import time
import pandas as pd
import numpy as np
import csv

output = [[]]

ser = serial.Serial('COM5', 9600, timeout=0.1) 
print("start")

for i in range(1000):

    data = ser.readline().decode('utf8')
    data_array= data.split(",")
    print(data_array)
    output.append(data_array)
    #time.sleep(0.001)

ser.close()

output = pd.DataFrame(output)
output.to_csv("freq_data.csv")
print(output)
print("end")
```
